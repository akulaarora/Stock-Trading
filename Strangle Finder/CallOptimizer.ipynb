```python
import tdameritrade as td
import pandas
import datetime
import os
```

```python
client_id = os.getenv('TDAMERITRADE_CLIENT_ID')
account_id = os.getenv('TDAMERITRADE_ACCOUNT_ID')
refresh_token = os.getenv('TDAMERITRADE_REFRESH_TOKEN')

td_client = td.client.TDClient(client_id=client_id, refresh_token=refresh_token, account_ids=[account_id])
```

```python
today = datetime.datetime.today()
seven_days = (today + datetime.timedelta(days=0)).strftime('%Y-%m-%d')
three_hundred_days = (today + datetime.timedelta(days=300)).strftime('%Y-%m-%d')
```

```python
ticker = 'FB'
quote = td_client.quoteDF(symbol=ticker)
price = (quote.at[0, 'askPrice'] + quote.at[0, 'bidPrice'])/2
df = td_client.optionsDF(symbol=ticker, includeQuotes=True, fromDate=seven_days, toDate=three_hundred_days, strikeCount=100)
df = df.loc[df['delta'] > 0]

df[['description', 'delta', 'vega', 'theta', 'ask', 'bid']]
df['mid'] = (df['ask']+df['bid'])/2
df[['description', 'delta', 'vega', 'theta', 'ask', 'bid', 'mid']]
df['gain'] = 1.5*df['delta'] + df['theta'] - 3*df['vega']
df['loss'] = -df['delta'] + df['theta'] - 3*df['vega']
df['% gain'] = 100*df['gain']/df['mid']
df['% loss'] = 100*(df['loss'])/df['mid']
df['leverage ratio'] = df['delta']*price/df['mid']
df[['description', 'delta', 'vega', 'theta', 'mid', 'gain', 'loss', '% gain', '% loss', 'leverage ratio']].sort_values('% gain', ascending=False)
```
